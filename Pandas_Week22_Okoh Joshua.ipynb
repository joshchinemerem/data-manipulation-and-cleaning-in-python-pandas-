# Group By and Aggregation Functions in Pandas

```python
import pandas as pd
```

```python
df =pd.read_csv(r"C:\Users\HP\Downloads\Flavors.csv")
df
```

```python
group_by_frame = df.groupby('Base Flavor')
```

```python
group_by_frame.mean('Base Flavor')
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').sum()
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating': {'mean', 'max', 'min', 'sum'}, 'Texture Rating': {'mean', 'max', 'min', 'sum'}})
```

```python
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating': {'mean', 'max', 'min', 'sum'}})
```

```python
df.groupby('Base Flavor').describe()
```
